```python
from justwatch import JustWatch

just_watch = JustWatch(country='US')

results = just_watch.search_for_item(title_id='603')
results
```

```python
movies = []
for item in results['items']:
    try:
        for thing in item['offers']:
            try:
                movies.append(thing['urls']['standard_web'])
            except KeyError:
                continue
    except KeyError:
        continue
movies = list(set(movies))
movies
```
